```python
import pandas
```

```python
errors = pandas.read_csv('data/all_errors.csv')
```

```python
# Drop collumns without numbers
errors = errors.dropna(axis=1, how='all')
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib
```

```python
# show up graphs
%matplotlib inline
```

```python
errors.mean()
```

```python
mean_errors = errors.mean()
```

```python
mean_plot = mean_errors.plot.bar(title="Mean Relative Error")
mean_plot.set_xlabel("Backpack size")
mean_plot.set_ylabel("Relative error")

plt.savefig('graphs/mean_errors', bbox_inches='tight')
```

```python
max_errors = errors.max()
```

```python
max_plot = max_errors.plot.bar(title="Max Relative Error")
max_plot.set_xlabel("Backpack size")
max_plot.set_ylabel("Relative error")
plt.savefig('graphs/max_errors', bbox_inches='tight')
```
