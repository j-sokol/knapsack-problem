```python
import pandas
```

```python
speeds = pandas.read_csv('data/all_times.csv')
```

```python
#Drop collumns without numbers
speeds = speeds.dropna(axis=1, how='all')
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib
```

```python
# Zapnout zobrazování grafů (procento uvozuje „magickou” zkratku IPythonu):
%matplotlib inline
```

```python
brute_speeds = [col for col in speeds.columns if 'brute' in col]
```

```python
brute_plot = speeds[brute_speeds].plot(title="Computation Speeds")
brute_plot.set_xlabel("Backpack ID")
brute_plot.set_ylabel("Time [s]")
plt.savefig('graphs/brute_force_speeds', bbox_inches='tight')
```

```python
heuristic_speeds = [col for col in speeds.columns if 'heuristic' in col]
```

```python
heuristic_plot = speeds[heuristic_speeds].plot(title="Computation Speeds")
heuristic_plot.set_xlabel("Backpack ID")
heuristic_plot.set_ylabel("Time [s]")
plt.savefig('graphs/heuristic_speeds', bbox_inches='tight')
```
